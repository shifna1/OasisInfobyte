# **Exploratory Data Analysis**

```python
# Finding the outliers of the 'Mntotal'
plt.figure(figsize = (6,4))
sns.boxplot(data = data, y = 'MntTotal')
plt.title('MntTotal Boxplot')
plt.ylabel('MntTotal')
plt.show()
```

```python
Q1 = data['MntTotal'].quantile(0.25)
Q3 = data['MntTotal'].quantile(0.75)

IQR = Q3 - Q1

low_bound = Q1 - 1.5 * IQR
up_bound = Q3 + 1.5 * IQR

outliers = data[(data['MntTotal'] < low_bound) | (data['MntTotal'] > up_bound)]
outliers.head()
```

```python
#removing Outliers
data = data[(data['MntTotal'] > low_bound) & (data['MntTotal'] < up_bound)]
data.describe()
```

```python
# Encoding the format and combine it so it will only in one column for each features.
data['marital_Married'] = data['marital_Married'].replace(1, 1)  # Keep 'Married' as 1
data['marital_Single'] = data['marital_Single'].replace(1, 2)
data['marital_Together'] = data['marital_Together'].replace(1, 3)
data['marital_Widow'] = data['marital_Widow'].replace(1, 4)
data['marital_Divorced'] = data['marital_Divorced'].replace(1, 5)

# Combine marital statuses into a single column
data['marital_Status'] = (
    data['marital_Married'] +
    data['marital_Single'] +
    data['marital_Together'] +
    data['marital_Widow'] +
    data['marital_Divorced']
)

# Map numeric values to descriptive marital statuses
marital_map = {
    1: 'married',
    2: 'single',
    3: 'together',
    4: 'widow',
    5: 'divorced'
}
data['marital_Status'] = data['marital_Status'].map(marital_map)
```

```python
cols_demographics = ['Income','Age']
```

```python
data.drop(['marital_Married','marital_Single','marital_Divorced','marital_Together','marital_Widow'], axis=1, inplace=True)
data['marital_Status'].unique()
```

```python
data['education_Basic'] = data['education_Basic'].replace(1,2)
data['education_Graduation'] = data['education_Graduation'].replace(1,3)
data['education_Master'] = data['education_Master'].replace(1,4)
data['education_PhD'] = data['education_PhD'].replace(1,5)
data['education_2n Cycle'] = data['education_2n Cycle'].replace(1,1)

data['education_Status'] = (data['education_2n Cycle']+data['education_Basic']+data['education_Graduation']+data['education_Master']+data['education_PhD'])
education_map = {
    1: '2n cycle',
    2: 'basic',
    3: 'graduation',
    4: 'master',
    5: 'phd'
}
data['education_Status'] = data['education_Status'].map(education_map)
```

```python
data.drop(['education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD'],axis = 1, inplace=True)
data['education_Status'].unique()
```

```python
def get_relationship(row):
    if row['marital_Status'] in ['married', 'together']:
        return 1
    else:
        return 0
data['In_Relationship'] = data.apply(get_relationship, axis = 1)
data.head()
```

# **K-means Clustering**

```python
# Standardising Data
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
clustering = ['Income', 'MntTotal', 'In_Relationship']
data_std = data.copy()
data_std[clustering] = std.fit_transform(data[clustering])
data_std[clustering].describe()
```

# **Elbow Method**

```python
X = data_std[clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters = K, random_state =7,  n_init=10).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10),inertia_list, color = (54 / 255, 113 / 255, 130 / 255))
plt.scatter(range(2, 10), inertia_list, color='red', marker='o', label='Cluster Points')
plt.title('Inertia vs. Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
data['Cluster'] = KMeans(n_clusters=5, random_state=7, n_init=10).fit_predict(X)

plt.figure(figsize=(8,6))
sns.scatterplot(x='MntTotal', y='Frequency', data=data, hue='Cluster', palette='viridis')
plt.title('Clustered Data Visual')
plt.xlabel('MntTotal')
plt.ylabel('Frequency')
plt.show()
```
